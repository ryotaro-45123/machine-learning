```python
#Import libraries
from sklearn.decomposition import NMF
from sklearn.datasets import make_blobs
```

```python
#Make data
centers = [[5, 10, 5], [10, 4, 10], [6, 8, 8]]
V, _ = make_blobs(centers=centers)
```

```python
#Apply model
n_components = 2
model = NMF(n_components = n_components)
model.fit(V)
W = model.transform(V)
H = model.components_
print(W)
print(H)
```
